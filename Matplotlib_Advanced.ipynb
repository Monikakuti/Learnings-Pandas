```python
#🌟 Advanced Features in Matplotlib 
'''Feature    | Real-Life Analogy       | Keyword
Legend        | Name tags               | plt.legend()
Gridlines     | Lined notebook          | plt.grid(True)
Annotation    | Sticky notes            | plt.annotate()
Style         | Instagram filters       | plt.style.use()
Save Plot     | Screenshot of your work | plt.savefig()
Multiple Axes | Comic book frames       | plt.subplots()
3D Plotting   | Turning map to globe    | projection='3d' '''
```

```python
# 1️⃣ Legends – Like Name Tags at a Party
# Imagine you're at a party, and people are wearing name tags to identify themselves. 
# Similarly, in a plot with multiple lines or bars, legends help us know which line represents what.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 2, 5]
y2 = [2, 3, 5, 7]

plt.plot(x, y1, label='Team A', color='blue')
plt.plot(x, y2, label='Team B', color='green')

plt.legend()  # Shows the name tags
plt.title('Team Performance')
plt.show()

# 🧠 Memory Tip: Think “legend = label guide”
```

```python
# 2️⃣ Gridlines – Like Lined Notebooks
# When you plot something on paper, gridlines help you align values easily — just like notebook lines 🗒️.
# Example:
plt.plot([1, 2, 3], [2, 4, 6])
plt.grid(True)  # Adds notebook-like lines
plt.title('With Gridlines')
plt.show()

#📌 Useful for reading values precisely.
```

```python
#3️⃣ Annotations – Like Sticky Notes on a Chart
# Imagine you're presenting a chart and want to point at a value and say, 
# “Look here!” That’s what annotate() does — like a sticky note pointing to something important.
# Example:
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)

# Add a note pointing to a value
plt.annotate('Peak', xy=(4, 30), xytext=(3, 32),
             arrowprops=dict(facecolor='red', shrink=0.05))

plt.title('Sales Trend')
plt.show()

#🧠 Memory Trick: annotate = add a note
```

```python
#4️⃣ Styles – Like Instagram Filters for Your Plots
# Matplotlib has built-in styles like ggplot, seaborn, or dark_background that make your plots instantly prettier.
# Example:
plt.style.use('ggplot')  # Apply a built-in style

plt.plot([1, 2, 3], [3, 5, 7])
plt.title('Styled Plot')
plt.show()

#📸 It’s like applying a filter on your data story!
```

```python
#5️⃣ Saving Plots – Like Taking a Screenshot
# You can save your chart as an image file (PNG, JPG, PDF). It’s like taking a snapshot of your work.
# Example:
plt.plot([1, 2, 3], [2, 4, 6])
plt.title('Save Me Please!')
plt.savefig('my_chart.png')  # Saves the file
plt.show()

#🧠 Pro Tip: Always plt.show() after savefig() to avoid blank images.
```

```python
#6️⃣ Multiple Axes / Subplots – Like a Comic Strip Panel
# You can draw multiple plots side by side — like different frames in a comic book 📖.
# Example:
fig, (ax1, ax2) = plt.subplots(1, 2)  # 1 row, 2 columns

ax1.plot([1, 2, 3], [3, 5, 7])
ax1.set_title('Line')

ax2.bar(['A', 'B', 'C'], [5, 7, 6])
ax2.set_title('Bar')

plt.suptitle('Multiple Axes')
plt.show()

#🧠 Think of fig as the canvas, ax1, ax2 as frames inside.
```

```python
#7️⃣ 3D Plots – Like Turning a 2D Map into a 3D Globe
#You can add a 3D twist to your plots using mpl_toolkits.mplot3d.
# Example:
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)
y = np.sin(x)
z = np.cos(x)

ax.plot3D(x, y, z, 'green')
plt.title('3D Line Plot')
plt.show()

#🧠 Use when you have 3 features or dimensions to show.
```
