```python
#ü•ä Seaborn vs. Matplotlib ‚Äî Cheat Sheet
'''Feature          | Matplotlib                 | Seaborn
üì¶ Purpose         | Base plotting library      | Built on top of Matplotlib, high-level
üé® Style           | Manual styling             | Beautiful defaults (colors, fonts)
üî¢ Works well with | Lists, NumPy arrays        | Pandas DataFrames
üß† Complexity      | More code for pretty plots | Less code, smarter visuals
üìà Plot types      | Line, bar, scatter, etc.   | Plus: violin, box, heatmap, pairplot
üßº Clean output    | Needs tuning               | Cleaner out of the box
üåà Color palettes  | Manual                     | Built-in palette functions'''

# Matplotlib = ‚Äúraw paintbrush üé®‚Äù
# Seaborn = ‚Äúready-made templates & themes üìä‚Äù

#üí° Real-Life Analogy for Seaborn
# Matplotlib is like painting a wall with brushes and rollers manually.
# Seaborn is like using peel-and-stick designer wallpaper ‚Äî fast, neat, and stylish!
```

```python
'''Plot Type  | Use Case
lineplot()    | Time series, trends
barplot()     | Categorical comparisons
boxplot()     | Spread, outliers
violinplot()  | Spread + distribution
scatterplot() | Correlation, patterns
pairplot()    | Full variable relationship grid
heatmap()     | Correlation or matrix data'''
```

```python
# Import
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#1Ô∏è‚É£ Line Plot ‚Äì Like a Temperature Chart Over Days
# # Sample Data

data = sns.load_dataset('flights')  # built-in dataset
data = data.pivot(index='month', columns='year', values='passengers')

# Plot
sns.heatmap(data, cmap='YlGnBu')  # color-coded passenger numbers
plt.title('Monthly Flight Passengers')
plt.show()

#üß† Memory Tip: Seaborn = Smart + Stylish
```

```python
#2Ô∏è‚É£ Bar Plot ‚Äì Like a Scoreboard
tips = sns.load_dataset('tips')  # restaurant tips dataset

sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Avg Bill by Day')
plt.show()

#üìå You only need to say what data to use and Seaborn does the rest.
```

```python
#3Ô∏è‚É£ Box Plot ‚Äì Like a Data Thermometer
# Shows min, max, median, and outliers ‚Äî great for comparing spread.

sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Bill Range by Day')
plt.show()

#üß† Think of this like measuring how wide and wild the bills are.
```

```python
#4Ô∏è‚É£ Violin Plot ‚Äì Box Plot + Distribution = üéª

sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Bill Spread by Day (Violin)')
plt.show()

#Looks cooler and gives more insight into shape of the data.
```

```python
#5Ô∏è‚É£ Scatter Plot with Hue ‚Äì Like Adding Color to Categories

sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)
plt.title('Tip vs Bill (by Gender)')
plt.show()

#Color (hue) splits data by categories.
```

```python
#6Ô∏è‚É£ Pair Plot ‚Äì All-in-One Relationship Matrix
sns.pairplot(tips, hue='sex')
plt.suptitle('Relationships Between Features')
plt.show()
# It auto-plots every feature against every other. Great for EDA.
```

```python
#7Ô∏è‚É£ Heatmap ‚Äì Like a Weather Map of Values

# Load dataset
tips = sns.load_dataset('tips')

# Select only numeric columns
numeric_tips = tips.select_dtypes(include='number')

# Correlation and heatmap
corr = numeric_tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

#Colors show strength of relationships. Red = strong pos, Blue = strong neg.
```
